```python
import pandas as pd
import warnings
import mysql.connector
import os
from dotenv import load_dotenv

# Onderdruk de UserWarning
warnings.simplefilter("ignore", UserWarning)

```

<div style="background-color:#006069; color:white; padding: 10px;">
    <strong>Server Variabelen met .env</strong>
</div>




```python
load_dotenv()
db_host = os.getenv('DB_HOST')
db_port = os.getenv('DB_PORT', 3306)
db_user = os.getenv('DB_USER')
db_password = os.getenv('DB_PASSWORD')
db_name = os.getenv('DB_NAME')
db_ssl_ca = os.getenv('DB_SSL_CA')
```

<div style="background-color:#006069; color:white; padding: 10px;">
    <strong>Bekijken Databases</strong>
</div>




```python
# Connect to the database
conn = mysql.connector.connect(
    host=db_host,
    port=db_port,
    user=db_user,
    password=db_password,
    database=db_name,
    ssl_ca=db_ssl_ca  
)

#Query the database
query = "SHOW Databases"
databases = pd.read_sql(query, conn)

# Close the connection
conn.close()

# Display the data  
databases
```

<div style="background-color:#006069; color:white; padding: 10px;">
    <strong>Bekijken Tabellen</strong>
</div>




```python
# Connect to the database
conn = mysql.connector.connect(
    host=db_host,
    user=db_user,
    password=db_password,
    database=db_name,
    ssl_ca=db_ssl_ca  
)

#Query the database
query = "SHOW Tables"
tables = pd.read_sql(query, conn)

# Close the connection
conn.close()

# Display the data
tables
```

<div style="background-color:#006069; color:white; padding: 10px;">
    <strong>Loop voor het snel bekijken van alle tabellen achter elkaar</strong>
</div>

```python
# Establish the connection to Azure MySQL
conn = mysql.connector.connect(
    host=db_host,
    user=db_user,
    password=db_password,
    database=db_name,
    ssl_ca=db_ssl_ca  
)

# Haal alle tabellen op
show_tables_query = "SHOW TABLES"
tables_df = pd.read_sql(show_tables_query, conn)

# Loop door alle tabellen en haal gegevens op
for table in tables_df.iloc[:, 0]:
    query = f"SELECT * FROM {table}"
    df = pd.read_sql(query, conn)
    
    # Toon een voorbeeld van de gegevens per tabel
    print(f"Data van tabel: {table}")
    display(df.head(10))  # Eerste 10 rijen kan aanpassen naar een ander getal
    print("\n")

# connectie uitzettenm
conn.close()
```
